##Step2: Read Image from Urls

> In this step we will read images from urls, and display them using openCV

```python
 image = io.imread("/content/drive/MyDrive/DATA for colab/sid.jpg") 
 cv2_imshow(image)
```

##Step3: Operation performed on the Image


```python
grey_filter = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(grey_filter)
# grey scale image generated from real image
```

```python
invert = cv.bitwise_not(grey_filter)
cv2_imshow(invert)
# grey scale image is inverted
```

```python
blur= cv.GaussianBlur(invert, (21,21), 0)
cv2_imshow(blur)
# inverted image is blured 
#cv2.gaussianblur() function to apply Gaussian Smoothing on the input source image
#any sharp edges in images are smoothed
```

```python
invertedblur= cv.bitwise_not(blur)
cv2_imshow(invertedblur)
# blur image is inverted
```

```python
sketch_filter= cv.divide (grey_filter, invertedblur, scale=255.0)
# using inverted blur image and grey filter image combination sketch image is being created
cv2_imshow(sketch_filter)
cv.imwrite("sketch.jpg", sketch_filter)
```
